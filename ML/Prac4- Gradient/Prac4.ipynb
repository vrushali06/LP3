```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
def objective(x):
    return (x+3)**2
```

```python
def derivative(x):
    return 2*(x+3)
```

```python
def gradient(alpha,start,max_itr):
    x_list=[]
    x=start
    x_list.append(x)
    for _ in range(max_itr):
        grad=derivative(x)
        #update x
        x=x-(alpha*grad)
        x_list.append(x)
    return x_list
```

```python
alpha=0.1
start=2
max_itr=30

arr=gradient(alpha,start,max_itr)
```

```python
#objective ==>> corr 
x_cor=np.linspace(-5,5,100)
```

#plot

```python
#plot curve line
plt.plot(x_cor,objective(x_cor))
#plot point
plt.plot(2,objective(2),'ro')

#plot actual local minima
x_arr=np.array(arr)
plt.plot(x_arr,objective(x_arr),'.-',color='black')
```
